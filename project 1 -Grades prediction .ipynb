<h1 style="background-color:#ABFF9B;font-size:20px;color:#00033E;font-weight : bold">✅ Importing Required Libraries</h1>

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.utils import shuffle


data = pd.read_csv("dataset for grade prediction.csv")
data.describe()    # Dataset description
```

<h1 style="background-color:#ABFF9B;font-size:20px;color:#00033E;font-weight : bold">✅ To check any null values present in dataset</h1>

```python
data.isnull().any()   
```

<h1 style="background-color:#ABFF9B;font-size:20px;color:#00033E;font-weight : bold">✅Prepare the data  and process using Linear Regression </h1>

**The G3 column is target label, split the dataset into 20% testing and 80% training.** 


```python
data = data[["G1", "G2", "G3", "studytime", "failures", "absences"]]
predict = "G3"
x = np.array(data.drop([predict], 1))
y = np.array(data[predict])

from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2)

linear_regression = LinearRegression()
linear_regression.fit(xtrain, ytrain)
```

<h1 style="background-color:#ABFF9B;font-size:20px;color:#00033E;font-weight : bold"> ✅Accuracy score for LR :</h1> 

```python
accuracy = linear_regression.score(xtest, ytest)
print(accuracy)
```

<h1 style="background-color:#ABFF9B;font-size:20px;color:#00033E;font-weight : bold">✅ predictions by the  model:</h1> 



```python
predictions = linear_regression.predict(xtest)
for i in range(len(predictions)):
    print(predictions[x], xtest[x], [ytest[x]])
```
